Las Series son estructuras unidimensionales indexadas con etiquetas. Es decir que cada elemento tiene una etiqueta asociada que permite identificarlo y acceder al elemento.
Estas etiquetas son llamadas "Indice".

### 3.2. Funciones lambda

Son funciones anónimas, es decir que no hace falta definirlas explicitamente con **def** donde usualmente realizamos acciones simples sobre algún argumento. Su potencial se ve cuando se usa en conjunto con otras funciones como lo veremos en unos momentos.

Una funcion normal luce así:

```python
def funcion_normal(dataVolume):
  return dataVolume * 2

datos = 13.4

print("Resultado de función normal:",datos, "x 2 =", funcion_normal(datos))
```

La misma función expresada como lambda sería asi:

```python
funcion_lambda = lambda dataVolume: dataVolume * 2

datos = 13.4

print("Resultado de función lambda:", datos, "x 2 =", funcion_lambda(datos))
```

### 3.3. Apply()

La función lambda por si sola puede parecer un poco sin sentido pero su magia ocurre cuando se combina con otros metodos, como el método **apply()**

Supongamos que de nuestra salida anterior, queremos generar una nueva columna cuyo valor sean los primeros dos caracteres del campo celda. Una manera de lograr esto podría ser:

```python
df_local_1 = celdas_general
df_local_1['abreviacion'] = df_local_1['celda'].str[:2]
df_local_1
```

Otra manera de hacerlo, con el metodo **apply()**, podria ser:

```python
df_local_2 = celdas_general
df_local_2['abreviacion'] = df_local_2.apply(lambda row: row['celda'][:2], axis=1)
df_local_2.head()
```

En este caso la operación a realizar es simple, sobre un string y se resuelve rápidamente con un accessor. La diferencia se nota en casos donde necesitamos aplicar alguna lógica, por ejemplo:

**Si deseo generar un nuevo campo que sea igual a AMAMBAY si las primeramos dos letras de la columna nodo es AM y en caso contrario que sea "DESCONOCIDO".**

```python
df_local_3 = celdas_general
df_local_3['ciudad'] = df_local_3.apply(lambda row: 'AMAMBAY' if row['nodo'][:2] == 'AM' else 'DESCONOCIDO', axis=1)
df_local_3.head()
```
