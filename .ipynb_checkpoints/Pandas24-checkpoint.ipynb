

## Un ejemplo más complicado:

* **Natural join:** para mantener solo las filas que coinciden con los marcos de datos, especifique el argumento `how = 'inner'`.
* **Full outer join:** para mantener todas las filas de ambos dataframe, especifique `how = 'OUTER'`.
* **Left outer join:** para incluir todas las filas de su dataframe x y solo aquellas de y que coincidan, especifique `how= 'left'`.
* **Right outer join:** para incluir todas las filas de su dataframe y y solo aquellas de x que coincidan, especifique `how=   
 'right'`.

```python
resultado=provincia.merge(hw_simple, left_on='CODIGO', right_on='Site', how='inner' )
```

```python
smod_counts = resultado.groupby('PROVINCIA_NOMBRE')['SMOD'].count()

# Contar las ocurrencias de BBMOD por provincia
bbmod_counts = resultado.groupby('PROVINCIA_NOMBRE')['BBMOD'].count()
```

```python
plt.figure(figsize=(12, 6))
smod_counts.plot(kind='bar')
plt.title('Cantidad de SMOD por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Cantidad de SMOD')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
bbmod_counts.plot(kind='bar')
plt.title('Cantidad de BBMOD por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Cantidad de BBMOD')
plt.xticks(rotation=90)
plt.show()
```

```python
contingency_table = pd.crosstab(resultado['PROVINCIA_NOMBRE'], resultado['SMOD'])

# Contar las ocurrencias de BBMOD por provincia
```

```python
contingency_table.plot(kind='bar', stacked=True, figsize=(12, 8))

# Añadir títulos y etiquetas
plt.title('Tipos de SMOD por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Cantidad de SMOD')
plt.xticks(rotation=90)
plt.legend(title='SMOD')

# Mostrar el gráfico
plt.show()
```

GRAFICO DE TORTA
