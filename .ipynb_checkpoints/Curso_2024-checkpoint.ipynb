## GRAFICOS con PANDAS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))

sns.countplot(x='class', data=titanic)
# Añadir títulos y etiquetas
plt.title('Cantidad de Pasajeros por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad de Pasajeros')
plt.show()
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sns.countplot(x='class', hue='survived', data=titanic)

# Añadir títulos y etiquetas
plt.title('Cantidad de Sobrevivientes por Clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad de Pasajeros')
plt.legend(title='Sobrevivió', labels=['No', 'Sí'])

# Mostrar el gráfico
plt.show()
```

```python
# Crear el scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', hue='class', data=titanic)

# Añadir títulos y etiquetas
plt.title('Precio del Boleto vs. Edad')
plt.xlabel('Edad')
plt.ylabel('Precio del Boleto')
plt.legend(title='Clase')

# Mostrar el gráfico
plt.show()
```

```python
titanic['class'].value_counts().plot(kind='bar')      
```

```python
geo = pd.read_csv('geo.csv',low_memory=False )
```

```python
geo = geo[['CODIGO','LATITUD' , 'LONGITUD']]
```

```python
geo
```

```python
gsm = gsm[['Site','BSC_Name']]
```

```python
gsm.drop_duplicates(inplace=True)
```

```python
gsm
```

```python
gsm_geo= gsm.merge(geo, how="inner" , left_on="Site" , right_on="CODIGO")
```

```python
hw_gsm=hw[hw['TYPE']=='GSM SA']
```

```python
hw_gsm = hw_gsm[['Site', 'SMOD','BBMOD' ]] 
```

```python
hw_gsm
```

```python
gsm_geo_df=gsm_geo.merge(hw_gsm , how="inner", on='Site' )
```

```python
import geopandas as gpd
```

```python
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
geometia = [Point(xy) for xy in zip (gsm_geo_df['LONGITUD'],gsm_geo_df['LATITUD'])]
```

```python
gsm_c = gpd.GeoDataFrame(gsm_geo_df , geometry=geometia)
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
# Plotear el mapa base y los puntos
fig, ax = plt.subplots(figsize=(10, 10))
world.plot(ax=ax, color='lightgrey')
gsm_c.plot(ax=ax, color='red', markersize=50)

# Añadir títulos y etiquetas
plt.title('Puntos en el Mapa')
plt.xlabel('Longitud')
plt.ylabel('Latitud')

# Mostrar el gráfico
plt.show()
```
